# Seaborn

[**Seaborn**](https://seaborn.pydata.org/#) is another Python data visualization library that is built on top of Matplotlib. Seaborn’s interface makes it an attractive choice for generating statistical plots. 

Furthermore, its close integration with Pandas’ `DataFrame` data structure makes it a frequently used data visualization library

```python
# import seaborn
import seaborn as sns

sns.get_dataset_names()
```

```python
# penguins dataset

penguins_df = sns.load_dataset('penguins')
penguins_df
```

```python
penguins_df.isnull().sum()
```

```python
# drop NaN
#penguins_df.info()

penguins_df = penguins_df.dropna()
penguins_df
```

```python
penguins_df.isnull().sum()
```

The types of graphs you want to use depends on what data you have and which you want to visualize
For simple categorical data a bar chart can do the trick


**Categorical Plots**

Categorical plots are used to visualize *categorical* data. This type of data is organized in distinct groups (Cities, Countries, Species, etc) 

```python
# Categorical 
sns.barplot(x='island',y='flipper_length_mm',data=penguins_df);
```

```python
# Multiple Categorical

sns.catplot(data=penguins_df,kind='bar',x='island',y='flipper_length_mm',hue='sex');
# swarm, violin, box, boxen
```

```python
# semi-colon not necessary
sns.catplot(data=penguins_df,x='island',y='flipper_length_mm',hue='sex');
```

```python
sns.catplot(data=penguins_df,x='island',y='flipper_length_mm',hue='sex',col='species');
```

Say you have quantitative now (numerical values)

**Histograms**

Histograms are used to visualize the distribution of one or more variables of our data. The data is organized in (discrete) bins and the height of each *bar* varies depending on the number (count) of observations that are found within each bin

```python
# Quantitative

sns.histplot(data=penguins_df,x='flipper_length_mm');
# sns.histplot(data=penguins_df,y='flipper_length_mm');
# Horizontal
```

```python
sns.histplot(data=penguins_df,x='flipper_length_mm', bins=5, binwidth=3);
```

```python
sns.histplot(data=penguins_df,x='flipper_length_mm', bins=5, binwidth=3, shrink=0.8);
#binwidth overides bins argument
#shrink leaves a space between the bins
```

```python
sns.histplot(data=penguins_df,x='flipper_length_mm', bins=5, binwidth=3, shrink=0.8, hue='sex', multiple='stack', fill=False);
```

```python
sns.histplot(data=penguins_df,x='flipper_length_mm', bins=5, hue='sex');
```

```python
sns.histplot(data=penguins_df,x='flipper_length_mm', bins=5, hue='sex', multiple='stack');
```

**Scatterplots**

Scatterplots allow us to visualize the relationships between data points. They show us how different features are related to each other 

```python
# Multi Quantitative

sns.scatterplot(data=penguins_df,x='flipper_length_mm', y='bill_length_mm');
```

```python
markers={"Male": "+", "Female": "x"}
sns.scatterplot(data=penguins_df,x='flipper_length_mm', y='bill_length_mm',hue='species', style='sex', markers=markers);
```

```python
# Multi Quantitative w/ 1 Categorical
sns.scatterplot(data=penguins_df,x='flipper_length_mm', y='bill_length_mm',hue='sex');
```

**Boxplots**

Boxplots or Box and Whisker plots are used to visualize the distribution through *quartiles*. The box represents the quartiles while the whiskers extend to illustrate the rest of the distribution. Points that are considered to be outliers are depictred by the diamond shaped dots

The line in the middle of the box represents the **Median** or Q2

```python
# 1 Categorical and 1 Quantitative 
sns.boxplot(data=penguins_df,y='flipper_length_mm', x='island');
```

**Heatmaps**

Heatmaps are used to visualize data that is organized in a 2-dimensional matrix. Heatmaps plot data as a color-encoded matrix

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month','year','passengers')
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights, annot=True, fmt='d')
```

```python
ax = sns.heatmap(flights, annot=True, fmt='d', cmap='Greens')
```
