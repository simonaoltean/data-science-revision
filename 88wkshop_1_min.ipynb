```python
# 1.1 Select unique values in the year column and order them in acsending order. How many unique values are there?
pd.read_sql_query("SELECT DISTINCT(year), COUNT(year) as count FROM results GROUP BY year ORDER BY year asc", conn)
```

```python
# 1.3 If a vehicle is older than 45 years, then label it as "45+". 
# Otherwise, label it as "<45". Create a new column that takes these two values
pd.read_sql_query("SELECT *, CASE WHEN 123-year <45 THEN '<45' ELSE '45+' END AS age_grp FROM results",conn)
```

```python
# 1.4 How many vehicles are in our dataset that are under 45 and 45 and over?
pd.read_sql_query("""SELECT COUNT(*) AS count,
                    CASE WHEN 123-year <45 THEN '<45' ELSE '45+' END AS age_grp FROM results GROUP BY age_grp"""
                  ,conn)
```

```python
# 2.2.1 Use a single query to see if your favorite song is in this dataset
pd.read_sql_query("""SELECT * FROM song_data WHERE song_name LIKE '%Purple Haze%'""", conn)
```

```python
# 2.2.2 How many unique popularity scores are there?
pd.read_sql_query("""SELECT COUNT(DISTINCT song_popularity) FROM song_data""", conn)
```

```python
# 2.2.3 How many song names are in this dataset that contain 'radio edit' in their name? 
# Make sure your query is not case sensitive and there is only one single character (space) between 'radio' and 'edit'.
pd.read_sql_query("""SELECT COUNT(song_name) FROM song_data WHERE LOWER(song_name) LIKE '%radio edit%'""",conn)
```

```python
# 2.2.4 Write an SQL query to report songs with popularity of 99 and above. How many songs are there?
pd.read_sql_query("""SELECT DISTINCT(song_name) FROM song_data WHERE song_popularity>99""",conn)
```

There is one song, Happier.

```python
# 2.2.5 Create a new column that contains song duration in the following format m:s. 
# Example, the duration of Boulevard of Broken Dreams is 4:22
pd.read_sql_query("""SELECT song_name, song_duration_ms,
           (song_duration_ms / 60000) || ':' || strftime('%S', (song_duration_ms / 1000), 'unixepoch') AS time_in_m_s
            FROM song_data
            """,conn)
```

```python
# 2.2.6 Display the minimum, maximum and average popularity scores. 
# The output of your query should contain only one row with 3 columns.
pd.read_sql_query("""SELECT MIN(song_popularity) AS minimum, MAX(song_popularity) AS maximum,
                    ROUND(AVG(song_popularity),2) AS average FROM song_data""", conn)
```

```python
# 2.2.7 Display average popularity scores and count for songs that contain "love" in their name and those that don't. 
# Is there any difference in their popularity?
pd.read_sql_query("""SELECT ROUND(AVG(song_popularity),2) AS average FROM song_data WHERE song_name LIKE '%love%'""",conn)
```

```python
pd.read_sql_query("""SELECT ROUND(AVG(song_popularity),2) AS average FROM song_data WHERE song_name NOT LIKE '%love%'""",conn)
```

Yes, the songs that contain 'love' in their name score a higher average song popularity of 54.39 than the ones that don't, with a score of 52.94.
