# 2.2 Queries

If you want more practice, explore built-in [String Functions](https://www.sqlitetutorial.net/sqlite-string-functions/) and [Aggregate Functions](https://www.sqlitetutorial.net/sqlite-aggregate-functions/) using the songs dataset.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as pld
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
song = pd.read_csv('../data/song_data.csv')
conn = sqlite3.connect('song.sqlite')
song.to_sql('song_data',con=conn,index=False,if_exists='replace')
song = pd.read_csv('../data/song_data.csv')
conn = sqlite3.connect('song.sqlite')
pd.read_sql_query("""SELECT * FROM song_data""",conn)
```

#### SUBSTR()

```python
pd.read_sql_query("""SELECT SUBSTR(song_name, 0, 2) FROM song_data""", conn)
```

#### TRIM()

```python
pd.read_sql_query("""SELECT TRIM(song_name) FROM song_data""", conn)
```

#### LTRIM()

```python
pd.read_sql_query("""SELECT LTRIM(song_name) FROM song_data""", conn)
```

#### RTRIM()

```python
pd.read_sql_query("""SELECT RTRIM(song_name) FROM song_data""", conn)
```

#### LENGTH()

```python
pd.read_sql_query("""SELECT LENGTH(song_name) FROM song_data""", conn)
```

#### REPLACE()

```python
pd.read_sql_query("""SELECT REPLACE(song_name, 'y', 'yyy') AS 'wayyy too much' FROM song_data""", conn)
```

#### UPPER()

```python
pd.read_sql_query("""SELECT UPPER(song_name) FROM song_data WHERE song_name LIKE '% loving %'""", conn)
```

#### LOWER()

```python
pd.read_sql_query("""SELECT LOWER(song_name) FROM song_data WHERE song_name LIKE '%love%'""", conn)
```

#### INSTR()

```python
pd.read_sql_query("""SELECT song_name, INSTR(song_name, 'Love') POSITON FROM song_data WHERE song_name LIKE '%love%'""", conn)
```

#### Concatenation Operator ||

```python
pd.read_sql_query("""SELECT TRIM(ROUND(tempo),'.0') || ' BPM' AS 'BPM' FROM song_data""", conn)
```

#### AVG()

```python
pd.read_sql_query("""SELECT RTRIM(ROUND(AVG(danceability),2),".0") AS 'danceability for songs containing "dance"' 
                    FROM song_data WHERE song_name LIKE '%dance%'""", conn)
```

```python
pd.read_sql_query("""SELECT RTRIM(ROUND(AVG(danceability),2),".0") AS 'danceability for songs containing "crying"' 
                    FROM song_data WHERE song_name LIKE '%crying%'""", conn)
```

#### COUNT()

```python
pd.read_sql_query("""SELECT COUNT(*) AS 'songs containing "dance"' FROM song_data WHERE song_name LIKE '%dance%'""", conn)
```

```python
pd.read_sql_query("""SELECT COUNT(*) AS 'songs containing "cry"' FROM song_data WHERE song_name LIKE '%cry%'""", conn)
```

#### MAX()

```python
pd.read_sql_query("""SELECT MAX(tempo) FROM song_data""", conn)
```

#### MIN()

```python
pd.read_sql_query("""SELECT MIN(energy) FROM song_data""", conn)
```

#### SUM()

```python
keys=pd.read_sql_query("""SELECT key, SUM(song_popularity) AS popularity FROM song_data GROUP BY key""", conn)
keys
```

```python
sns.barplot(x='key',y='popularity',data=keys)
plt.title('Song Popularity by Key')
plt.show()
```

#### GROUP_CONCAT()

```python
concat_name=pd.read_sql_query("""SELECT GROUP_CONCAT(DISTINCT(song_name)) AS concat_name FROM song_data WHERE song_popularity>95""", conn)
concat_name
```

```python
print(concat_name.concat_name.values)
```

```python
df = pd.read_sql_query("""SELECT song_popularity, CASE WHEN tempo<60 THEN '0-60' 
                        WHEN tempo>=60 and tempo<80 THEN '60-80' 
                        WHEN tempo>=80 and tempo<100 THEN '80-100'
                        WHEN tempo>=100 and tempo<120 THEN '100-120'
                        WHEN tempo>=120 and tempo<140 THEN '120-140'
                        WHEN tempo>=140 and tempo<160 THEN '140-160'
                        WHEN tempo>=160 and tempo<180 THEN '160-180'
                        WHEN tempo>=180 THEN '180+'
                        END AS tempo_grp FROM song_data""", conn)
df
```

```python
df=df.groupby(['tempo_grp']).mean()
df=df.reindex(['0-60','60-80','80-100','100-120','120-140','140-160','160-180','180+'])
df['tempo_group']=df.index
df
```

```python
sns.barplot(data=df, x='tempo_group', y='song_popularity', palette='mako')
plt.title('Song Popularity by Tempo')
plt.show()
```
